### Question 1. Plotting graphs For IRIS Dataset Using Seaborn Library And matplotlib.pyplot library 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv("Iris.csv")
iris.head()
```

### A.Creating and customizing line charts using Matplotlib 

```python
iris.drop(['Species','Id'], axis=1).plot.line(title='Iris Dataset')
```

### B. Visualizing relationships between two or more variables using scatter plots 

```python
# create a figure and axis
fig, ax = plt.subplots()
# scatter the sepal_length against the sepal_width
ax.scatter(iris['SepalLengthCm'], iris['SepalWidthCm'])

ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}

fig, ax = plt.subplots()

for i in range(len(iris['SepalLengthCm'])):
    ax.scatter(iris['SepalLengthCm'][i], iris['SepalWidthCm'][i],color=colors[iris['Species'][i]])

ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

### C. Studying distributions of variables using histograms & bar charts

```python
fig, ax = plt.subplots()

data = iris['Species'].value_counts()

points = data.index
frequency = data.values

ax.bar(points, frequency)

ax.set_title('Species Frequency')
ax.set_xlabel('Species')
ax.set_ylabel('Frequency')
```

```python
fig, ax = plt.subplots()

data = iris['SepalLengthCm'].value_counts()

points = data.index
frequency = data.values

ax.bar(points, frequency)

ax.set_title('Iris Dataset')
ax.set_xlabel('SepalLength')
ax.set_ylabel('Frequency')
```

### D. Visualizing two-dimensional data using heatmaps 

```python
sns.heatmap(iris.corr(), annot=True)
```

### E. Plotting multiple Matplotlib and Seaborn charts in a grid

```python
sns.pairplot(iris)
```
